```python
variables = [
    "contract",
    "gender",
    "payment_method",
    "internet_service",
    "paperless_billing"
]
```

```python
import pandas as pd
```

```python
def churn_proportion_table(df, column):

    table = pd.crosstab(df[column], df["churn"], normalize="index") * 100

    table_long = (
        table
        .reset_index()
        .melt(id_vars=column, var_name="churn", value_name="proportion")
    )

    # Solo churn = 1
    table_plot = table_long[table_long["churn"] == 1]

    return table_plot
```

```python
contract_plot = churn_proportion_table(telcom_data, "contract")
gender_plot = churn_proportion_table(telcom_data, "gender")
payment_plot = churn_proportion_table(telcom_data, "payment_method")
internet_plot = churn_proportion_table(telcom_data, "internet_service")
paper_plot = churn_proportion_table(telcom_data, "paperless_billing")
```

```python
# Identifying numeric categories :

telcom_data.select_dtypes(include = "number").columns
```

```python
telcom_data['tenure_group'] = pd.cut(
    telcom_data['tenure'],
    bins=[0, 12, 24, 48, 100],
    labels=['0-12', '13-24', '25-48', '49+']
)
```

```python
# Total customers by tenure group

total_by_range = (
    telcom_data.groupby('tenure_group', observed = True).size().reset_index(name='total')
)

total_by_range
```

```python
# Churned customers by tenure group

churn_by_range = (
    telcom_data[telcom_data['churn']== 1].groupby('tenure_group', observed= True).size().reset_index(name='churn'))
churn_by_range
```

```python
#Merge totals and churn

tenure_churn = total_by_range.merge(churn_by_range, on='tenure_group', how ='left')
```

```python
# replace nan with 0

tenure_churn['churn'] = ternure_churn['churn'].fillna(0)

#calculate churn percentage

tenure_churn['churn_rate'] = (ternure_churn['churn'] / ternure_churn['total'] *100)

tenure_churn
```

```python
tenure_churn['tenure_group'] = tenure_churn['tenure_group'].astype(str)
```

```python
# Count tenure Evasion and graph

plt.figure(figsize=(10, 6))

ternure_graph = sns.barplot( x= 'tenure_group', y = 'churn', data = tenure_churn, palette = 'Greens', hue = 'tenure_group')
plt.xlabel('tenure group', labelpad= 15)
plt.ylabel('Count')
plt.title('Churn rate by tenure (count)')

for p in ternure_graph.patches:
  ternure_graph.text(p.get_x() + p.get_width()/2,
                      p.get_height() /2,
                      f'{int(p.get_height())}',
                      ha = 'center',
                      va = 'center',
                      fontsize = 12,
                      color = 'black')

ternure_graph.set_yticks([])
ternure_graph.spines['top'].set_visible(False)
ternure_graph.spines['right'].set_visible(False)
ternure_graph.spines['left'].set_visible(False)
ternure_graph.spines['bottom'].set_visible(False)


plt.show()
```
